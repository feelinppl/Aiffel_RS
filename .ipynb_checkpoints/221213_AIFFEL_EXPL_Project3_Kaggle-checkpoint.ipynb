## Guide

**<튜닝해볼 수 있는 모델 클래스 인자>**<br>
대표적으로 자주 튜닝하는 lightgbm 라이브러리의 인자는 다음과 같습니다.
- max_depth : 의사 결정 나무의 깊이, 정수 사용
- learning_rate : 한 스텝에 이동하는 양을 결정하는 파라미터, 보통 0.0001~0.1 사이의 실수 사용
- n_estimators : 사용하는 개별 모델의 개수, 보통 50~100 이상의 정수 사용
- num_leaves : 하나의 LightGBM 트리가 가질 수 있는 최대 잎의 수
- boosting_type : 부스팅 방식, gbdt, rf 등의 문자열 입력

위에서는 n_estimators와 max_depth 에 대해서만 아주 간단하게 실험했지만, 훨씬 더 다양하게 실험을 하며 최적의 조합을 찾아볼 수 있습니다.
실제로 lightgbm에 넣을 수 있는 인자는 아래 자료들에서 설명하는 것처럼 매우 다양합니다.

## 세 가지 이상의 다양한 모델에 대해 하이퍼 파라미터 튜닝하기 LGBMRegressor, XGBRegressor, RandomForestRegressor

- boosting algorithm vs bagging algorithm
- [What is difference between bagging and boosting](https://quantdare.com/what-is-the-difference-between-bagging-and-boosting/)

```python
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
def rmse(y_test, y_pred):
    return np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(y_pred)))
```

```python
train = concat_df_new.iloc[:train_index,2:]
X_final_test = concat_df_new[train_index:]
X_final_test = X_final_test.drop(['id', 'date'], axis = 1)
print(train.shape, X_final_test.shape, y.shape)
```

```python
random_state=42        
gboost = GradientBoostingRegressor(random_state=random_state)
xgboost = XGBRegressor(random_state=random_state)
lightgbm = LGBMRegressor(random_state=random_state)
rdforest = RandomForestRegressor(random_state=random_state)

models = [gboost, xgboost, lightgbm, rdforest]
```

```python
# for loop and seaborn subplots: https://stackoverflow.com/questions/53310228/how-can-i-make-seaborn-distribution-subplots-in-a-loop
# feature_importances_plotting: https://www.boostcourse.org/ds214/lecture/102113?isDesc=false
scores = {}
random_state=42 
fig,axes = plt.subplots(2,2, figsize = (15,8))
plt.suptitle("Models & RMSE")

for model, ax in zip(models, axes.flat):
    model_name = model.__class__.__name__

    X_train, X_test, y_train, y_test = train_test_split(train, y, random_state=random_state, test_size=0.2)

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # 예측 결과의 rmse값 저장
    scores[model_name] = rmse(y_test, y_pred)
    
    # data frame에 저장
    # score_df = pd.DataFrame(scores, index=['RMSE']).T.sort_values('RMSE', ascending=False)
    sns.barplot(x = model.feature_importances_, y = X_train.columns, ax = ax)
    ax.set_title(model_name + ', score: ' + '{:.2f}'.format(rmse(y_test, y_pred)))
plt.show()
```

- LGBM Regressor가 가장 낮은 RMSE점수를 보이는 것으로 확인됨
- 대부분의 model에서 grade가 가장 중요한 역할을 하는 것으로 보이며 LGBMRegressor에서만 latitude와 longitude가 가장 중요한 영향을 미친 것으로 보임
- 반면, bedrooms나 bathrooms, 그리고 yr_renovated등은 집값에 큰 영향을 미치지 않는 것으로 보임

```python
params = {}

param_grid_gb = {'max_depth': range(4,10),
             'max_features': [0.3,0.5,0.75,1],
             'n_estimators': [200, 500, 1000],
             'learning_rate': np.random.uniform(0.01, 0.2, 5)
             }

param_grid_xgb = {'max_depth': range(4,10),
                  'n_estimators': [200, 500, 1000],
                  'learning_rate': np.random.uniform(0.01, 0.2, 5)
                  }

param_grid_lgbm = {'max_depth': range(4,10),
             'num_leaves': [15,30,45,60],
             'n_estimators': [200, 500, 1000],
             'learning_rate': np.random.uniform(0.01, 0.2, 5),
             }

param_grid_rf = {'max_depth': range(4,10),
             'n_estimators': [200, 350, 500],
             'max_leaf_nodes': [None, 10, 20, 30],
             }
params['param_grid_gb'] = param_grid_gb
params['param_grid_xgb'] = param_grid_xgb
params['param_grid_lgbm'] = param_grid_lgbm
params['param_grid_rf'] = param_grid_rf
```

## Baseline 커널에서 활용했던 블렌딩 방법 활용하기
- Average Blending 여러 가지 모델의 결과를 산술평균하여 블렌딩 모델
- Grid Search로 찾은 하이퍼 파라미터 값 적용

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
random_state=42     
gboost = GradientBoostingRegressor(random_state=random_state, learning_rate = 0.0820, max_depth=5, max_features=0.75, n_estimators=17000)
xgboost = XGBRegressor(random_state=random_state, learning_rate = 0.0725, max_depth = 5, n_estimators = 1700)
lightgbm = LGBMRegressor(random_state=random_state, learning_rate = 0.0732, max_depth=8, n_estimators= 1700, num_leaves = 15)
rdRegressor = RandomForestRegressor(random_state=random_state, max_depth=9, max_leaf_nodes=None, n_estimators = 500)

models_base = [{'model':gboost, 'name':'GradientBoosting'}, 
          {'model':xgboost, 'name':'XGBoost'},
          {'model':lightgbm, 'name':'LightGBM'},
          {'model':rdRegressor, 'name':'rdRegressor'}]
```

### Cross validation

```python
def get_cv_score(models, X_train, y):
    kfold = KFold(n_splits=5).get_n_splits(X_train.values)
    for m in models:
        CV_score = np.mean(cross_val_score(m['model'], X=X_train.values, y=y, cv=kfold))
        print(f"Model: {m['name']}, CV score:{CV_score:.4f}")
```

```python
get_cv_score(models_base, train, y)
```

- 모델의 cv 스코어 확인 결과 GradientBoosting, XGBoost, LightGBM모델을 블렌딩에 사용해 보기로함
- 이 후 더 나은 스코어를 줄 수 있는 조합을 바꿔가면서 적용해보기로 함

### Averaging Blending

```python
def AveragingBlending(models, x, y, sub_x):
    for m in models : 
        m['model'].fit(x.values, y)
    
    predictions = np.column_stack([
        m['model'].predict(sub_x.values) for m in models
    ])
    return np.mean(predictions, axis=1)
```

```python
y_pred = AveragingBlending(models_base[0:3], train, y, X_final_test)
print(len(y_pred))
np.exp(y_pred)
```
